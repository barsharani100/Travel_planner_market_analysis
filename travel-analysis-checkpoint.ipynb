```python
# load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
# Load the provided CSV file
file_path = 'Top Indian Places to Visit.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print(data.head())
print(data.info())
print(data.describe())
```

## Assuming the correct column names are as follows:
## 'Time needed to visit (in hours)', 'Google review rating', 'Entrance Fee (in INR)', 'Number of Google reviews (in lakhs)'


```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Handle missing values
data = data.dropna()  #dropping missing values

# Display the column names to identify the correct names
print(data.columns)
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Handle missing values(imputation strategy)
data = data.dropna()  #dropping missing values
```

```python
# Encoding categorical variables
data['Zone'] = data['Zone'].astype('category').cat.codes
data['State'] = data['State'].astype('category').cat.codes
data['City'] = data['City'].astype('category').cat.codes
data['Type'] = data['Type'].astype('category').cat.codes
data['Significance'] = data['Significance'].astype('category').cat.codes
```

```python
# Check if 'DSLR allowed' column exists
if 'DSLR allowed' in data.columns:
    data['DSLR allowed'] = data['DSLR allowed'].apply(lambda x: 1 if x == 'Yes' else 0)
else:
    print("Column 'DSLR allowed' not found.")
```

```python
# Check if 'Airport within 50km radius' column exists
if 'Airport within 50km radius' in data.columns:
    data['Airport within 50km radius'] = data['Airport within 50km radius'].apply(lambda x: 1 if x == 'Yes' else 0)
else:
    print("Column 'Airport within 50km radius' not found.")
```

```python
# Check if 'Weekly off' column exists
if 'Weekly off' in data.columns:
    data['Weekly off'] = data['Weekly off'].apply(lambda x: 0 if x == 'None' else 1)
else:
    print("Column 'Weekly off' not found.")
```

```python
# Columns to be normalized, adjust based on actual columns available
columns_to_normalize = ['Time needed to visit (in hours)', 'Google review rating', 'Entrance Fee (in INR)', 'Number of Google reviews (in lakhs)']
existing_columns_to_normalize = [col for col in columns_to_normalize if col in data.columns]

if existing_columns_to_normalize:
    scaler = StandardScaler()
    data[existing_columns_to_normalize] = scaler.fit_transform(data[existing_columns_to_normalize])
else:
    print("No columns found to normalize.")

print(data.head())
```

```python
# Plot the distribution of Google review ratings
plt.figure(figsize=(10, 6))
sns.histplot(data['Google review rating'], bins=20, kde=True)
plt.title('Distribution of Google Review Ratings')
plt.xlabel('Google Review Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of entrance fees
plt.figure(figsize=(10, 6))
sns.histplot(data['Entrance Fee in INR'], bins=20, kde=True)
plt.title('Distribution of Entrance Fees')
plt.xlabel('Entrance Fee (INR)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the relationship between entrance fee and Google review rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Entrance Fee in INR', y='Google review rating', data=data)
plt.title('Entrance Fee vs Google Review Rating')
plt.xlabel('Entrance Fee (INR)')
plt.ylabel('Google Review Rating')
plt.show()
```

```python
# Analyze the number of popular destinations per state
popular_destinations = data.groupby('State')['Google review rating'].count().reset_index()
popular_destinations.columns = ['State', 'Count']
plt.figure(figsize=(10, 6))
sns.barplot(x='State', y='Count', data=popular_destinations)
plt.title('Number of Popular Destinations per State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```
